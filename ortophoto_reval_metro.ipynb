### rm 21-0.05

```python
r = pd.DataFrame()
for rok in df['akt_rok'].unique():
    x = df[df['akt_rok'] == rok].groupby('piksel').count()[['akt_rok', 'kolor', 'modul_archiwizacji']]
    x['rok'] = rok
    r = pd.concat([r, x])
r
```

```python
df = df.where((df['piksel'] != 0.05) | (df['akt_rok'] != 2021))
df
```

```python
df = df.dropna(subset=['akt_rok', 'url_do_pobrania'])
df
```
